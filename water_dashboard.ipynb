```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# TITLE
# -------------------------
st.title("🌊 River Flow Forecasting System")

# -------------------------
# USER INPUTS
# -------------------------
river = st.selectbox(
    "Select River System",
    ["Colorado", "Mississippi", "Hudson"]
)

horizon = st.slider(
    "Forecast Horizon (hours)",
    1, 24, 6
)

# -------------------------
# RIVER DATA MAPPING
# -------------------------
DATA_PATHS = {
    "Colorado": "/Users/afrahboateng/Documents/DATASCI 207 Applied Machine Learning/Water Project/data/raw/passaic_combined_v2.csv",
    "Mississippi": "/Users/afrahboateng/Documents/DATASCI 207 Applied Machine Learning/Water Project/data/raw/passaic_combined_v2.csv",
    "Hudson": "/Users/afrahboateng/Documents/DATASCI 207 Applied Machine Learning/Water Project/data/raw/passaic_combined_v2.csv",
    "Passaic": "/Users/afrahboateng/Documents/DATASCI 207 Applied Machine Learning/Water Project/data/raw/passaic_combined_v2.csv"
 }

# -------------------------
# LOAD DATA BASED ON USER INPUT
# -------------------------
file_path = DATA_PATHS[river]

df = pd.read_csv(file_path)
df["datetime"] = pd.to_datetime(df["datetime"])
df = df.set_index("datetime")
df = df.sort_index()

# -------------------------
# SIMPLE MOCK FORECAST (replace with your model later)
# -------------------------
last_value = df["Discharge (ft^3/s)"].iloc[-1]

forecast = np.repeat(last_value, horizon)

future_index = pd.date_range(
    start=df.index[-1],
    periods=horizon+1,
    freq="h"
)[1:]

forecast_df = pd.DataFrame({
    "forecast": forecast
}, index=future_index)

# -------------------------
# PLOT
# -------------------------
fig, ax = plt.subplots(figsize=(10,4))

ax.plot(df.index[-100:], df["Discharge (ft^3/s)"].tail(100), label="History")
ax.plot(forecast_df.index, forecast_df["forecast"], label="Forecast")

ax.set_title(f"{river} River Flow Forecast")
ax.set_ylabel("Discharge")
ax.legend()

st.pyplot(fig)

# -------------------------
# OUTPUT
# -------------------------
st.write("Forecast Table")
st.dataframe(forecast_df)
```
